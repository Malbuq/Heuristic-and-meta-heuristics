```python
import matplotlib.pyplot as plt 
```

```python
from main import GA_algorithm
from FitnessEvaluator import SphereEvaluator, RosenbrockEvaluator, RastriginEvaluator
from FitnessSelector import FitnessTournamentSelector, FitnessProportionalSelector
from PopulationCrossover import OnePointCrossover, TwoPointsCrossover
```

```python
def plot_cenario(X,Y, evaluator, selector, mutation_probability, crossover_probability, crossover_type):
    plt.figure()
    plt.plot(X,Y)
    plt.title(f'Convergence graphic for {type(evaluator).__name__} and {type(selector).__name__} mutation: {mutation_probability} crossover prob: {crossover_probability} with {type(crossover_type).__name__}')
    plt.xlabel('Iterations')
    plt.xlabel('Best fitness')
    plt.show()
```

```python
NUMBER_ITERATIONS = 1000
LEFT_BOUND = -100
RIGHT_BOUND = 100
DIMENSION = 30
POPULATION_SIZE = 100

crossover_probabilities = [0.7, 0.8, 0.9]
mutation_probabilities = [0.01, 0.05, 0.1]
evaluators = [SphereEvaluator(), RastriginEvaluator(), RosenbrockEvaluator()]
selectors = [FitnessProportionalSelector(), FitnessTournamentSelector()]
crossovers = [OnePointCrossover(), TwoPointsCrossover()]
```

```python
def avarage(lst): 
    return sum(lst) / len(lst) 

def execute_X_times(X_TIMES, RIGHT_BOUND, POPULATION_SIZE, NUMBER_ITERATIONS, DIMENSION, crossover_probability, mutation_probability, evaluator, selector, crossover_type):
    Y_results = []
    X_results = []

    for i in range(X_TIMES):
        X, Y = GA_algorithm(RIGHT_BOUND, POPULATION_SIZE, NUMBER_ITERATIONS, DIMENSION, crossover_probability, mutation_probability, evaluator, selector, crossover_type)
        if i == 0:
            X_results = X

        Y_results.append(Y)

    Y_mean = []

    for i in range(len(Y_results[0])):
        Y_values = [curr_result[i] for curr_result in Y_results]

        Y_mean.append(avarage(Y_values))
    
    print(Y_results)
    print(Y_mean)
    return X_results, Y_mean
```

```python
for crossover_probability in crossover_probabilities:
    for mutation_probability in mutation_probabilities:
        for evaluator in evaluators:
            for selector in selectors:
                for crossover_type in crossovers:
                    X, Y = execute_X_times(10, RIGHT_BOUND, POPULATION_SIZE, NUMBER_ITERATIONS, DIMENSION, crossover_probability, mutation_probability, evaluator, selector, crossover_type)
                    plot_cenario(X,Y, evaluator, selector, mutation_probability, crossover_probability, crossover_type)
```
